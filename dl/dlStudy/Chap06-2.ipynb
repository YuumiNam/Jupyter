# 모두의 딥러닝 개정3판
### 06장 텐서플로우에서 실행하는 다중 선형 회귀

- 공부시간과 과외시간에 따른 성적의 변화를 선형회귀로 나타내어
- 임의의 공부시간을 입력했을때 성적이 얼만큼 나올지를 예측해봄
- 변수가 공부시간, 과외시간 2개이므로 다중선형회귀

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# 텐서플로의 케라스 API에서 필요한 함수들을 불러옴
from tensorflow.keras import models, layers

os.environ['KMP_DUPLICATE_LIB_OK']="TRUE"
```

```python
x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]]) # 공부시간, 과외시간
y = np.array([81, 93, 91, 97]) # 성정
```

```python
model = models.Sequential()

# 입력 변수가 두 개(공부시간, 과외시간)이므로 input_dim에 2를 입력
model.add(layers.Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer='sgd', loss='mse')
```

```python
# 오차를 최소화하는 과정을 1000번 반복
model.fit(x, y, epochs=1000)
```

```python
# 임의의 학습 시간과 과외 시간을 집어넣어 점수를 예측하는 모델을 테스트
hour = 7
private_class = 4
prediction = model.predict([[hour, private_class]])

print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점입니다." % (hour, private_class, prediction))
```
