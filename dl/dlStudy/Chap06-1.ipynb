### 06장 텐서플로우에서 실행하는 선형회귀

- 공부시간에 따른 성적의 변화를 선형회귀로 나타내어
- 임의의 공부시간을 입력했을때 성적이 얼만큼 나올지를 예측해봄

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# 텐서플로의 케라스 API에서 필요한 함수들을 불러옴
from tensorflow.keras import models, layers

os.environ['KMP_DUPLICATE_LIB_OK']="TRUE"
```

```python
x = np.array([2, 4, 6, 8]) # 공부 시간
y = np.array([81, 93, 91, 97]) # 점수
```

```python
model = models.Sequential()

# 출력 값, 입력 변수, 분석 방법에 맞게끔 모델을 설정
# 입력변수 = 공부시간
# 활성화함수 = 선형회귀
model.add(layers.Dense(1, input_dim=1, activation='linear'))
```

```python
# 오차 수정을 위해 경사 하강법(sgd)을 사용
# 평균 제곱 오차(mse)를 사용
model.compile(optimizer='sgd', loss='mse')
```

```python
# 오차를 최소화하는 과정을 1000번 반복
model.fit(x, y, epochs=1000)
```

```python
# 예측 결과를 그래프로 나타냄
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
# 임의의 시간을 집어넣어 점수를 예측하는 모델을 테스트
hour = 7
prediction = model.predict([hour])
print("%.f시간을 공부할 경우의 예상 점수는 %.02f점입니다." % (hour, prediction))
```
